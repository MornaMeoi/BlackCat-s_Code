```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

pic = plt.imread("KARTINKA.png")

middle = len(pic)//2
temp = np.rot90(pic[:middle, middle:, 0]).T.copy()
pic[:middle, middle:, 0] = np.flip(np.rot90(pic[:middle, :middle, 0])).T
pic[:middle, :middle, 0] = temp
pic[middle:, :middle, 0] = np.rot90(pic[middle:, :middle, 0]).T
pic[middle:, middle:, 0] = np.flip(pic[middle:, middle:, 0].T)

plt.figure(figsize=(15, 15))

res_1 = pic[:middle,:,0] + pic[:middle,:,2]
res_2 = pic[middle:,:,0] - pic[middle:, :, 2]

ul = np.flip(np.rot90((res_1[:, middle:]))).T
ur = np.rot90(res_1[:, :middle]).T
dl = np.rot90(res_2[:, :middle]).T
dr = np.flip(res_2[:, middle:]).T

res_1 = np.hstack((ul, ur)).reshape(256, 512)-1
res_2 = np.hstack((dl, dr)).reshape(256, 512)

plt.imshow(np.vstack((res_1[:, :], res_2[:, :])))
```
