```python
#task_84
def main():
    import random
    summ = 0
    for _ in range(10):
        counter = 0
        c0, c1 = 0, 0
        while c0 != 3 and c1 != 3:
            counter += 1
            if random.randint(0, 1) == 0:
                print("О", end=' ')
                c0 += 1
                c1 = 0
            else:
                print("Р", end=' ')
                c1 += 1
                c0 = 0
        print(f"(попыток: {counter})")
        summ += counter
    print(f"Среднее количество попыток: {summ/10}.")

if __name__ == "__main__":
    main()
```

```python
#task_109
def list_of_dates(first_year=1900, second_year=2000):
    list_of_dates = []
    days_in_months = [31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    for year in range(first_year, second_year): 
        for i in range(len(days_in_months)):
            for day in range(1, days_in_months[i]+1):
                list_of_dates.append(f"{day}/{i+1}/{year}")
            if i == 1 and year%4 == 0 and year%100 != 0:
                list_of_dates.append(f"29/2/{year}")
    return list_of_dates

def is_magic(date):
    date = date.split('/')
    if int(date[0]) * int(date[1]) == int(date[2][-2:]):
        print('/'.join(date))

def main():
    for date in list_of_dates():
        is_magic(date)

if __name__ == '__main__':
    main()
```
